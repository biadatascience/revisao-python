# Exercício 1) Aplicar descontos

Dada a seguinte tabela de produtos:

```python
import pandas as pd
```

```python
df = pd.DataFrame({'product_id': [1, 2, 3, 4, 5],
                   'name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Printer'],
                   'price': [1200, 20, 30, 200, 150],
                   'stock': [10, 50, 0, 15, 5]})
df.head()
```

Objetivo: Adicione uma nova coluna chamada discounted_price:

* Aplique 10% de desconto aos produtos com preço maior que 100.
* Para os demais produtos, mantenha o preço original.

```python
df['discounted_price'] = df['price']
```

```python
df.head()
```

```python
condicao = df['price'] > 100
```

```python
df.loc[condicao, 'discounted_price'] = df.loc[condicao, 'price']*0.9
```

```python
df.head()
```

# Exercício 2) Classificar notas dos alunos

Dada a tabela de alunos:

```python
df = pd.DataFrame({'product_id': [1, 2, 3, 4, 5],
                   'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
                   'grade': [85, 70, 90, 60, 75]
                   })
df.head()
```

Objetivo: Adicione uma coluna chamada status:

* "Pass" para alunos com nota maior ou igual a 75.
* "Fail" para alunos com nota menor que 75.

```python
df['status'] = 'Fail'
df.head()
```

```python
condicao = df['grade'] > 75
```

```python
df.loc[condicao, 'status'] = 'Pass'

df.head()
```

# Exercício 3) Categorizar temperatura

Dada a tabela de temperaturas:

```python
df = pd.DataFrame({'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
                   'temperature': [30, 15, 20, 40, 25]
                   })
df.head()
```

Objetivo: Adicione uma coluna chamada category:

* "Hot" para temperaturas maiores que 30.
* "Cold" para temperaturas menores que 20.
* "Moderate" para as demais.

```python
df['category'] = 'Moderate'
df.head()
```

```python
df.loc[df['temperature'] > 30, 'category'] = 'Hot'
df.loc[df['temperature'] < 20, 'category'] = 'Cold'
df
```

# Exercício 5) Classificar idades

Dada a tabela de pessoas:

```python
df = pd.DataFrame({'person_id': [1, 2, 3, 4, 5],
                   'name': ['John', 'Sara', 'Emily', 'Michael', 'Anna'],
                   'age': [25, 17, 65, 45, 10]})
df
```

Objetivo: Adicione uma coluna chamada age_group:

* "Child" para idades menores que 18.
* "Adult" para idades entre 18 e 60.
* "Senior" para idades maiores que 60.

```python
df['age_group'] = 'Child'
df
```

```python
condicao_adult = (df['age'] >= 18) | (df['age'] <= 60)
condicao_senior = df['age'] > 60
```

```python
df.loc[condicao_adult, 'age_group'] = 'Adult'
df.loc[condicao_senior, 'age_group'] = 'Senior'

df
```
