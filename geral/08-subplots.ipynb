# Subplots

Subplots permitem criar múltiplos gráficos em uma única figura. Isso é útil quando você quer comparar diferentes gráficos lado a lado ou em uma estrutura organizada.

O conceito básico de subplots envolve a criação de uma grade de gráficos dentro de uma mesma figura. Por exemplo, você pode criar uma grade de 2x2, onde você terá 4 gráficos organizados em duas linhas e duas colunas.

## Sintaxe básica

    plt.subplot(nrows, ncols, index)

Onde:

* nrows: número de linhas da grade

* ncols: número de colunas da grade

* index: posição do gráfico que você está criando (começa de 1)

# Exemplo 1) Utilizando o <i>subplot</i>

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Criando uma figura com 2 subplots (1 linha, 2 colunas)

plt.subplot(1, 2, 1)
plt.plot(x,y1)
plt.title('Seno')

plt.subplot(1, 2, 2)
plt.plot(x,y2)
plt.title('Cosseno')

plt.tight_layout() # Ajusta os subplots para que não se sobreponham

plt.show()
```

# Exemplo 2) Utilizando <i>subplots</i>

Vamos utilizar os mesmos dados anteriores, mas agora utilizaremos o método <i>subplots</i>

## Estrutura básica do <i>subplots</i>:

    fig, axes = plt.subplots(nrows, ncols)

Onde,

* nrows: número de linhas

* ncols: número de colunas

* axes: retorna um array de objetos de eixos (gráficos), que você pode acessar para plotar dados

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5))

# Primeiro subplot (Seno)
axes[0].plot(x, y1)
axes[0].set_title('Seno')

# Primeiro subplot (Seno)
axes[1].plot(x, y2, color='orange')
axes[1].set_title('Cosseno')

plt.show()
```

# Exercício 1) Criar dois subplots verticais

Crie uma figura com dois gráficos, um embaixo do outro. O gráfico superior deve mostrar a função seno e o inferior, a função cosseno.


```python
fig, ax = plt.subplots(2, 1, figsize=(5, 10))

# Criando o primeiro subplot
ax[0].plot(x, y1)
ax[0].set_title('Seno')

# Criando o segundo subplot
ax[1].plot(x, y1)
ax[1].set_title('Cosseno')

plt.show()
```

# Exercício 2)  Subplots com 4 gráficos

Crie uma grade de 2x2 com 4 gráficos. Cada gráfico deve mostrar uma função matemática diferente: seno, cosseno, tangente e uma linha reta.

```python
# Construindo a função tangente
y3 = np.tan(x)

# Construindo a linha reta
y4 = 3*x + 2

fig, ax = plt.subplots(2, 2, figsize=(6,6))

# Adicionando o primeiro plot na linha 0 e coluna 0
ax[0, 0].plot(x, y1)
ax[0, 0].set_title('Seno')

# Adicionando o segundo plot na linha 0 e coluna 1
ax[0, 1].plot(x, y2)
ax[0, 1].set_title('Cosseno')

# Adicionando o terceiro plot na linha 1 e coluna 0
ax[1, 0].plot(x, y3)
ax[1, 0].set_title('Tangente')

# Adicionando o quarto plot na linha 1 e coluna 1
ax[1, 1].plot(x, y4)
ax[1, 1].set_title('Reta')

# Ajustando o espaçamento entre os subplots
plt.subplots_adjust(wspace=0.4, hspace=0.4)  # Ajusta o espaço entre as colunas

plt.show()
```
