# Gráfico de pizza


<u><b>Definição</b></u>

Trata-se de um gráfico circular que ilustra parte ou partes de um todo. Todos os segmentos do gráfico somam 100%.

<u><b>Quando usar um gráfico de pizza?</b></u>

São usados para representar porcentagens, embora nem toda porcentagem que você encontre seja melhor visualizada com um gráfico de pizza. Ainda assim, o uso básico dessa ferramenta é indicado quando você deseja visualizar uma porcentagem ou partes de um todo.

<u><b>Quando NÃO usar um gráfico de pizza?</b></u>

A razão mais óbvia para não usar um gráfico de pizza é se você não tem porcentagens para ilustrar ou se as porcentagens não somam 100%. 

Além disso, os gráficos de pizza podem ficar difíceis de ler se você tiver mais do que algumas fatias. Idealmente, você deve limitar seu gráfico de pizza a não mais do que cinco fatias. Mesmo em gráficos de pizza que têm apenas algumas fatias, se alguma porcentagem estiver entre 5 e 10 pontos percentuais, será difícil para os leitores entenderem o que absorver da mensagem.

Tudo se reduz a isto: se o seu gráfico de pizza for muito complicado a ponto de não ser mais uma maneira eficaz de visualizar os dados, experimente um gráfico de barras ou gráfico de linhas. A visualização de dados não serve apenas para embelezar um design, então, se o método que você escolheu não ajuda as pessoas a visualizar as informações, escolha outro método.

```python
import matplotlib.pyplot as plt
```

# Exemplo 1) Exemplo simples da construção de um gráfico de pizza

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
```

```python
labels
```

```python
fig, ax = plt.subplots()
ax.pie(x=sizes, labels=labels)
plt.plot()
```

# Exemplo 2) Auto-label slices

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.2f%%')
plt.plot()
```
